```python
import pandas as pd
true_data = pd.read_csv('True.csv')
fake_data = pd.read_csv('Fake.csv')
```

```python
true_news = true_data.iloc[:,1]
fake_news = fake_data.iloc[:,1]
```

```python
#Natural Language Toolkit
import nltk
nltk.download('punkt')
```

```python
import matplotlib.pyplot as plt
```

```python
#nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
en_stops = stopwords.words('english')
```

```python
en_stops
```

```python
true_news_length = []
for each_true_news in true_news:
    tokens = nltk.word_tokenize(each_true_news)
    true_news_length.append(len(tokens))
    
true_news_length
```

```python
plt.hist(true_news_length, bins=20, histtype = 'step', label = 'True News')
plt.show()
```

```python
fake_news_length = []
for each_fake_news in fake_news:
    tokens = nltk.word_tokenize(each_fake_news)
    fake_news_length.append(len(tokens))
    
fake_news_length
```

```python
plt.hist(fake_news_length, bins=20, histtype = 'step', label = 'Fake News')
plt.show()
```

```python
plt.hist(true_news_length, bins=50, histtype = 'step', label = 'True News')
plt.hist(fake_news_length, bins=50, histtype = 'step', label = 'True News')
plt.show()
```

```python
stops = []
true_stops_length = []

for each_true_news in true_news:
    tokens = nltk.word_tokenize(each_true_news)
    stops.clear()
    for word in tokens: 
        if word in en_stops:
            stops.append(word)
    true_stops_length.append(len(stops))     
        
true_stops_length
```

```python
stops = []
fake_stops_length = []

for each_fake_news in fake_news:
    tokens = nltk.word_tokenize(each_fake_news)
    stops.clear()
    for word in tokens: 
        if word in en_stops:
            stops.append(word)
    fake_stops_length.append(len(stops))     
        
fake_stops_length
```

```python
plt.hist(true_stops_length, bins=50, histtype = 'step', label = 'True News')
plt.hist(fake_stops_length, bins=50, histtype = 'step', label = 'Fake News')
plt.show()
```
