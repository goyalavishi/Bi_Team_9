```python
#on single column 

import nltk
nltk.download()

import pandas as pd
import string

fakenews = pd.read_csv("Fake.csv")
truenews = pd.read_csv("True.csv")


fakenews =fakenews['text'][:500]
truenews = truenews['text'][:500]


fakenews=fakenews.str.lower()
truenews=truenews.str.lower()


#removing punctuations
exclude = set(string.punctuation)
def remove_punctuation(x):
    try:
        x = ''.join(ch for ch in x if ch not in exclude)
    except:
        pass
    return x
fakenews= fakenews.apply(remove_punctuation)
truenews= truenews.apply(remove_punctuation)


print("fakenews \n \n ",fakenews)
print("truenews \n \n ",truenews)
```

```python
from nltk.corpus import stopwords
#tokenisation and removing stopwords

fake_non_stops = []   #list of all stopwords from the fake news data
true_non_stops = []

en_stops=set(stopwords.words('english'))

for each_fake_news in fakenews:
    tokens=nltk.word_tokenize(each_fake_news)
    for word in tokens:
        if word not in en_stops:
            fake_non_stops.append(word)       
print(fake_non_stops)


for each_true_news in truenews:
    tokens=nltk.word_tokenize(each_true_news)
    for word in tokens:
        if word not in en_stops:
            true_non_stops.append(word)       
print("\n \n ",true_non_stops)


```

```python
#emotions algorithm
# NLP Emotion Algorithm   
# 1) Check if the word in the final word list is also present in emotion.txt
#  - open the emotion file
#  - Loop through each line and clear it
#  - Extract the word and emotion using split
# 2) If word is present -> Add the emotion to emotion_list
# 3) Finally count each emotion in the emotion list

fakenews_emotion_list = []
with open('emotions.txt', 'r') as file:
    for line in file:
        clear_line = line.replace("\n", '').replace(",", '').replace("'", '').strip()
        word, emotion = clear_line.split(':')
        if word in fake_non_stops:
            #print(word)
            fakenews_emotion_list.append(emotion)
print("\n \n fake news emotion list" , fakenews_emotion_list)


#true news emotionslist
truenews_emotion_list = []
with open('emotions.txt', 'r') as file:
    for line in file:
        clear_line = line.replace("\n", '').replace(",", '').replace("'", '').strip()
        word, emotion = clear_line.split(':')
        
        if word in true_non_stops:
            #print(word)
            truenews_emotion_list.append(emotion)
print("\n \n true news emotion list" , truenews_emotion_list)


```

```python
from collections import Counter

w=Counter(fakenews_emotion_list)
print("Fake News \n\n ",w)


v=Counter(truenews_emotion_list)
print("\n \n True news \n\n  ",v)
```

```python
import matplotlib.pyplot as plt

fig, ax1= plt.subplots()
ax1.bar(w.keys(), w.values()  )
fig.autofmt_xdate()
plt.savefig('FakeNewsgraph.png')
plt.xlabel('Emotions', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.title('Fake News',fontsize=20)
plt.show()

fig, ax1= plt.subplots()
ax1.bar(v.keys(), v.values() )
fig.autofmt_xdate()
plt.xlabel('Emotions', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.title('True News',fontsize=20)
plt.savefig('TrueNewsgraph.png')
plt.show()






```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
def sentiment_analyzer_method(text):
    score=SentimentIntensityAnalyzer().polarity_scores(text)
    print(score)
    neg=score['neg']
    pos=score['pos']
    
    if pos > neg :
        print("positive sentiment")
    elif neg > pos:
        print("negative sentiment")
    else:
        print("neutral sentiment")

sentiment_analyzer_method(fakenews[250])  #need non-tokenised text inside this function
sentiment_analyzer_method(truenews[250])

```
